## Filtering Data

```python
#And condition
df_csv.loc[(df_csv['Type 1'] == 'Grass') & (df_csv['Type 2'] == 'Poison')]
```

```python
# OR condition
df_csv.loc[(df_csv['Type 1'] == 'Grass') | (df_csv['Type 2'] == 'Poison')]
```

```python
# AND condition
df_csv.loc[(df_csv['Type 1'] == 'Grass') & (df_csv['Type 2'] == 'Poison') & (df_csv['Attack'] > 50) ]
```

```python
#Resetting index
new_df_csv = df_csv.reset_index()
```

```python
new_df_csv
```

```python
df_csv.loc[df_csv['Name'].str.contains('Mega')]
```

```python
# Does not contains
df_csv.loc[~df_csv['Name'].str.contains('Mega')]
```

```python
#  Filtering with RegEx

import re

df_csv.loc[df_csv['Type 1'].str.contains('fire|Grass', flags=re.I, regex=True)]
df_csv.loc[df_csv['Type 1'].str.contains('fire|Grass', regex=True)]
df_csv.loc[df_csv['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)]
df_csv.loc[df_csv['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

## Aggregate Statistics (Groupby)

```python
df_agr = pd.read_csv('/content/modified.csv')
df_agr
```

```python
#Mean based on a column value
df_agr.groupby(['Type 1']).mean()
df_agr.groupby(['Type 1']).mean().sort_values('Defense')

#Highest attack
df_agr.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df_agr.groupby(['Type 1']).sum()
```

```python
df_agr.groupby(['Type 1']).count()
```

```python
df_agr['count'] = 1
df_agr
#groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df_agr.groupby(['Type 1']).count()['count'].sort_values(ascending=False)
df_agr.groupby(['Type 2']).count()['count'].sort_values(ascending=False)
df_agr.groupby(['Type 1', 'Type 2']).count()['count'].sort_values(ascending=False)
```

## Working with llarge amouts of data

```python
for df_5rows in pd.read_csv('/content/modified.csv', chunksize=5):
    print("Chunk dataframe")
    print(df_5rows)
```

```python
new_df = pd.DataFrame(columns=df_5rows.columns)
```

```python
new_df
```

```python
for df_5rows in pd.read_csv('/content/modified.csv', chunksize=5):
    print("Chunk dataframe")
    results = df_5rows.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])
```
