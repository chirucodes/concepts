```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(seed=42)

df = pd.DataFrame(data=np.random.randint(0,100,(4,3)),
                  columns=['Test1','Test2', 'Test3'],
                  index=['bob','swift','galaxy','iphone'])
```

```python
df
```

### Only FALSE Values would be replaced.

```python
df<90
```

```python
df.where(df<90, 'A+')
```

```python
df.where(~(df<90), 'A+') #Wit negtion or to update where True values are there
```

```python
df
```

```python
df.where(df['Test2'] > 90, 'A+')  #Updating all the rows where 'Test2' value is <90
```

```python
df.where(df['Test2'] < 90, 'A+')
```

```python
df
```

```python
df.where(df['Test2']>90, 'A+', inplace=True)
```

```python
df
```

```python
np.random.seed(seed=42)

df = pd.DataFrame(data=np.random.randint(0,100,(4,3)),
                  columns=['Test1','Test2', 'Test3'],
                  index=['bob','swift','galaxy','iphone'])
df
```

```python
df.where(df['Test2']>90, lambda x: x+100, inplace=True)   #Apply lamda function where the condition is not satisfied
```

```python
df
```

```python
np.random.seed(seed=99)

another_df = pd.DataFrame(data=np.random.randint(0,100,(4,3)),
                  columns=['Test1','Test2', 'Test3'],
                  index=['bob','swift','galaxy','iphone'])
```

```python
another_df
```

### Getting the values form other dtaframe

```python
df.where(df['Test2']<90, another_df)
```

```python
df
```

### To update a specific cell in a dataframe using pandas.where

```python
df['Test1'] = df.where(df['Test1']>90,111)
```
