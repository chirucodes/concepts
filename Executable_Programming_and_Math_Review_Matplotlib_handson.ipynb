### Line Charts

Line Charts are used to understand the trend over time. It can explain the correlation between points by the trend.

Exercise 01: Plot a Line Chart for the given range values

```
import numpy as np
x1 = np.linspace(0, 5, 11)
print("x values:",x1)
y1 = x1 ** 2
print("y values:",y1)
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.plot(x1, y1, 'r--p', marker='o', markersize=10., label='xy')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.yticks([0,6,7,9,12,18,15,30])
plt.xticks([0,0.5,0.8,1,1.5,3,7])
plt.title('Plotting x and y values')
plt.show()
```

```
time = np.arange(0, 10, 0.5)
amplitude = np.sin(time)
c_amp = np.cos(time)
# print(time)
# print(amplitude)
plt.plot(time, amplitude, color="red", marker="o")
plt.show()
plt.plot(time, c_amp, color="yellow")
```

### Stack Plot

Stack plot visualizes data in stacks and shows the distribution of data over time. It is used for checking multiple variable area plots in a single plot.



Exercise 03: From the given lists sleeping, eating, working and playing hours on a scale of five days, perform a Stackplot.

```
days = [1,2,3,4,5]
sleeping =[7,8,6,11,7]
eating = [2,3,1,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
```

```
x = days
y = [sleeping, eating, working, playing]
plt.stackplot(x,y, labels = ['sleeping', 'eating', 'working', 'playing'])
plt.xticks([1,2,3,4,5])
plt.legend(loc='upper left') # bbox_to_anchor=(1,0.5)
plt.show()
```

### Bar Chart

Bar Chart gives the distribution of data over several groups

Exercise 04: For the given Apple iPhone Sales from 2012 to 2018, perform a Bar Plot with appropriate Title, X-label and Y-label.

```
salesInMillions = [125.05, 150.26, 169.22, 231.22, 211.88, 116.76, 217.72]
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
```

```
plt.title('Bar Plot')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.bar(years,salesInMillions, color=["red", "yellow", "green"])
plt.show()
```

```
colors = []
for each in salesInMillions:
    if each > 150:
        colors.append('green')
    else:
        colors.append('red')
plt.bar(years,salesInMillions, color=colors, width=.3)
plt.show()
```

Exercise 12: For the Given DataFrame, Perform the following plots along Two Rows and Two Columns:
* Line Plot with following measures,
  * marker = 'o'
* Line Plot with following measutes,
  * linestyle = 'none', marker='D', color="green"
* Scatter Plot with following measures,
  * c = df['y']
* Scatter Plot with following measures,
  * c = df['x']

```
import warnings
warnings.simplefilter("ignore")

df = pd.DataFrame({'x': range(1,101), 'y': np.random.randn(100)*15+range(1,101) })
print(df)

plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.plot( 'x', 'y', data=df, marker='o')

plt.subplot(222)
plt.plot( 'x','y', data=df, linestyle='none', marker='D', color="green")

plt.subplot(223)
plt.scatter( 'x','y', data=df, c = df['y']) #c is color mapping

plt.subplot(224)
plt.scatter( 'x','y', data=df, c = df['x'])
plt.show()
```

### Pie Chart

Exercise 13: For the given dictionary of IMDB Genres data, perform a Pie Plot with the following measures:
* Explode the 'Biography' Sliced section

Hint: [plt.pie](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pie.html)

```
IMDB_Genres = {
    'Action': 133,
    'Biography': 64,
    'Comedy': 267,
}
```

```
# explode is a len(x) array which specifies the fraction of the radius with which to offset each wedge.
explode = (0, 0.1, 0)
# autopct is a string or function used to label the wedges with their numeric value. 
plt.pie(IMDB_Genres.values(), autopct= '%1.1f%%' , explode=explode, labels = IMDB_Genres.keys(), shadow=True)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
